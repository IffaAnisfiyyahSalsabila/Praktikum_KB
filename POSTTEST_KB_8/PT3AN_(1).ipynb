<h1>Mencari Handling outlier (Menggunakan metode IQR)</h1>

```python
import pandas as pd
import numpy as np


file_path = '/content/top_10000_1960-now (1).csv'
df = pd.read_csv(file_path)

# Identify numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns


Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)


IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers in numeric columns
outliers = (df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound)

print("Number of outliers in each column:")
print(outliers.sum())

# Optionally, remove outliers from the DataFrame
df_cleaned = df[~outliers.any(axis=1)]

# Display the cleaned dataset (without outliers)
print("\nCleaned dataset preview:")
print(df_cleaned.head())
```

<h1>Membuat Feature engineering</h1>

```python
import pandas as pd

# Load the dataset
file_path = '/content/top_10000_1960-now (1).csv'
data = pd.read_csv(file_path)

# 1. Extracting Year from Album Release Date
data['Album Release Date'] = pd.to_datetime(data['Album Release Date'], errors='coerce')
data['Release Year'] = data['Album Release Date'].dt.year

# 2. Create an 'Audio Profile' by averaging audio features
audio_features = ['Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence']
data['Audio Profile'] = data[audio_features].mean(axis=1)

# 3. Word Count for Track Name, Artist Name, and Album Name
data['Track Name Word Count'] = data['Track Name'].apply(lambda x: len(str(x).split()))
data['Artist Name Word Count'] = data['Artist Name(s)'].apply(lambda x: len(str(x).split()))
data['Album Name Word Count'] = data['Album Name'].apply(lambda x: len(str(x).split()))

# 4. Categorize the release year into Decades
data['Decade'] = (data['Release Year'] // 10) * 10

# 5. Length of the Song Title (character count)
data['Track Name Length'] = data['Track Name'].apply(lambda x: len(str(x)))


output_file = '/content/top_10000_1960-now_featured.csv'
data.to_csv(output_file, index=False)


data[['Track Name', 'Release Year', 'Audio Profile', 'Track Name Word Count', 'Decade', 'Track Name Length']].head()
```
